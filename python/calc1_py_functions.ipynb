```python
# with multiple outputs
def awesomeFunction(in1,in2):
  sumres = in1+in2
  prodres = in1*in2
  print('Their sum is ' + str(sumres))
  print('Their product is ' + str(prodres))
  return sumres,prodres
```

```python
out1,out2 = awesomeFunction(4,5)

print(out1,out2)
```

```python
# lambda functions

littlefun = lambda x : x**2 - 1

littlefun(4)
```
