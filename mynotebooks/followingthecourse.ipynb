```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

## Polynomial functions

Create polynomials with numpy and sympy and create sine waves using polynomials (Maclaurin series)

```python
order = 4

xlimhigh = 5
xlimlow = -5

coefs_limhigh = 3
coefs_limlow = -3

np.random.seed(41)

coefs = (coefs_limhigh - coefs_limlow) * np.round(np.random.rand(order),2) + coefs_limlow
x = sym.symbols('x')
poly = sym.Poly(coefs, x)
plt.plot(np.linspace(xlimlow, xlimhigh, 101), [poly.subs(x, i) for i in np.linspace(xlimlow, xlimhigh, 101)])
plt.grid()
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
print(poly.as_expr())
plt.title('$Polynomial: \\quad %s$' % sym.latex(poly.as_expr()))
plt.show()
```
